```sql
--Show all employees from Sales department.
select * from employees where department='Sales'
```

```sql
--Show employees whose salary > 90000, sorted by salary descending
select * from employees where salary >90000 order by salary desc
```

```sql
--Show employees whose first_name starts with 'A'
select * from employees where first_name like 'A%'
```

```sql
--Show all distinct cities where employees work
select distinct city from employees
```

```sql
--Count how many employees are Active
select count(*) from employees where status='Active'
```

```sql
--Count how many distinct departments exist
select count(distinct department) from employees
```

```sql
--Count employees in HR or Finance departments
select count(*) from employees where department in('HR','Finance')
```

```sql
--Show the top 5 highest-paid employees
select * from employees order by salary desc limit 5
```

```sql
--Show 5 employees, skipping the first 3, sorted by hire_date ascending
select * from employees order by hire_date asc limit 5 offset 3
```

```sql
--Show Active employees in Hyderabad or Pune whose salary > 70000, sorted by department ascending, then salary descending, limit 3
select * from employees where status='Active' and city in ('Hyderabad','Pune') and salary >70000 order by department asc,salary desc limit 3
```

```sql
--Show distinct departments of employees whose salary > 80000, sorted alphabetically
select distinct department from employees where salary>80000 order by department asc
```

```sql
--Count employees whose first_name starts with 'S' and are Active
select count(*) from employees where status='Active' and first_name like'S%'
```
