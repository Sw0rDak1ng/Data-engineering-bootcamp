```sql
--Find the total number of employees in each department.
select department,count(*)as Total_Emp from employees group by department
```

```sql
--Find the average salary of employees in each department.
select department,avg(salary)as avg_salary from employees group by department
```

```sql
----Find the average salary of employees in each department (decimal 2 point)
select department,round(avg(salary),2) as avg_salary from employees group by department
```

```sql
--Count how many employees have each job title.
select job_title,count(*)as total_employees from employees group by job_title
```

```sql
--Find the number of employees living in each city.
select city,count(*) as place_live_employees from employees group by city
```

```sql
--Find departments where the average salary is greater than 50,000.
select department,avg(salary)  from employees group by department having avg(salary) > 50000
```

```sql
--Find the highest salary in each department.
select department,max(salary) from employees group by department
```

```sql
--Find the total salary expense per city (sum of salaries for all employees living in each city).
select city,sum(salary) from employees group by city 
```

```sql
--Count how many employees were hired on each date.
select year(hire_date) as hired_year,count(*) as total_employees from employees group by year(hire_date)
```
