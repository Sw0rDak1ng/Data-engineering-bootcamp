```sql
--Count total number of employees
select count(*) as total_employees from employees
```

```sql
--Count employees whose salary is greater than 80,000
select count(*) as salary_greater_than_80k from employees where salary >80000
```

```sql
--Count employees from the Engineering department
select count(*) as Total_employees_from_Eng_department from employees where department='Engineering'
```

```sql
--Count employees who are Active
select count(*) as active_employees from employees where status='Active'
```

```sql
--Count distinct departments
select count(distinct department) as distinct_department from employees
```

```sql
--Count distinct cities
select count(distinct city) as distinct_city from employees
```

```sql
--Count distinct job departments where salary > 70000
select count(distinct department) from employees where salary >70000
```

```sql
--Count employees from HR or Finance
select count(*) from employees where department in('HR','Finance')
```

```sql
--Count employees from Hyderabad who are Active
select count(*) from employees where status='Active' and city='Hyderabad'
```

```sql
--How many employees have a first name starting with 'S'?
select count(*) from employees where first_name like 'S%'
```

```sql
--Check how many employees have salary between 60000 and 90000
select count(*) from employees where salary between 60000 and 90000
```

```sql
--Count employees hired after 2020
select count(*) from employees where hire_date <'2020-01-01'
```
