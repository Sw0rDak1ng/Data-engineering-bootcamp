GROUP BY


```sql
select department from employees group by 1
```

```sql
--count the no of employees in the deparment 
select department,count(*) as Employees from employees group by 1
```

```sql
--List all departments that have fewer than 9 employees, along with the number of employees in each department.
select department,count(*) as employees from employees group by 1 having employees <9
```
