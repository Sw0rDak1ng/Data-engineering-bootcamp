```sql
--Show all employees whose last_name ends with 'son', sorted by salary descending.
select * from employees where last_name like'%son' order by salary desc
```

```sql
--Show the top 5 employees whose salary is between 60000 and 90000, sorted by hire_date ascending
select * from employees where salary between 60000 and 90000 order by hire_date asc limit 5
```

```sql
--Show distinct cities where employees from Engineering department work.
select distinct city from employees where department='Engineering'
```

```sql
--Count the number of Active employees whose first_name starts with 'R'.
select count(*)as ListofActiveEmployees from employees where status='Active' and first_name like'R%'
```

```sql
--Show 3 employees from HR or Finance with the highest salary, sorted by salary descending.
select * from employees where department in('HR','Finance') order by salary desc limit 3
```

```sql
--Show employees who are Active and work in Delhi or Pune, sorted by department ascending, salary descending.
select * from employees where status='Active' and city in('Delhi','Pune') order by department asc,salary desc
```

```sql
--Count the number of distinct departments that have employees with salary > 80000.
select count(distinct department) from employees where salary >80000 
```

```sql
--Show 3 employees whose first_name starts with 'S', skipping the first 2, sorted by salary descending.
select * from employees where first_name like'S%' order by salary desc limit 3 offset 2
```

```sql
--Count how many employees have a salary between 70000 and 100000.
select count(*) from employees where salary between 70000 and 100000
```

```sql
--Show Active employees from Engineering or Sales whose salary > 75000, sorted by city ascending, salary descending, limit 5.
select * from employees where department in('Engineering','Sales') and salary>75000 order by city asc,salary desc limit 5
```
