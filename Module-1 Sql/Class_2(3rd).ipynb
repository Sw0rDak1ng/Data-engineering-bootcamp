ORDER BY QUESTIONS

```sql
--Salary of the top person working in the organization
select * from employees order by salary desc
```

```sql
--salary of the lowest person working in the organization 
select * from employees order by salary asc
```

```sql
--showing the department name from desc( meaning from the last alphabetic order)
select * from employees order by department desc
```

```sql
--showing the city name from desc(last alphabetic order) and first_name from asc(first alphabetic order)
select * from employees order by city desc ,first_name asc
```

```sql
--showing the first_name asc
select * from employees order by first_name asc;
```

```sql
--showing the job titles from the ascending order
select * from employees order by job_title asc
```

```sql
--showing the city name from ascending order
select * from employees order by city asc
```

```sql
--recently hired person 
select * from employees order by hire_date desc
```

```sql
--recently hired person only the top one to show 
select * from employees order by hire_date desc limit 1
```

```sql
--find the first highest salary 
select * from employees order by salary desc limit 1
```

```sql
--find the second highest salary 
select * from employees order by salary desc limit 1 offset 1
```

```sql
--find the 3rd highest salary 
select * from employees order by salary desc limit 1 offset 2
```
