```sql
--Show a list of all unique departments.
select distinct department from employees
```

```sql
--Show all distinct cities where employees work.
select distinct city from employees
```

```sql
--Show distinct job titles in the company.
select distinct job_title from employees
```

```sql
--Show distinct combinations of department and city.
select distinct department,city from employees
```

```sql
--Show distinct salaries greater than 70000.
select distinct salary  from employees where salary >70000 order by salary desc
```

```sql
--Show distinct first_name values that start with ‘S’.
select distinct first_name from employees where first_name like'S%' order by first_name asc
```

```sql
--Count the number of distinct departments in the company.
select distinct department from employees count department
```
